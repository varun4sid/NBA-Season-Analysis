```python
import requests
from bs4 import BeautifulSoup
```

```python
link = "https://www.nba.com/stats/players/traditional"

page = requests.get(link)
page_soup = BeautifulSoup(page.text, "lxml")
print(page_soup)
```

```python
main_table = page_soup.find_all("table",{"class":"Crom_table__p1iZz"})

for table in main_table:
    with open("table.html","+a") as file:
        file.write(table)
```

```python
with open("table.html","r") as file:
    soup = BeautifulSoup(file, "lxml")
    rows = soup.find_all("tr")
```

```python
headers = []
for element in rows[0].find_all("th"):
    header = element.text.strip()
    headers.append(header)
    
headers.remove('')
headers
```

```python
result = []
index=1
for row in rows[1:]:
    record = [index]
    for element in row.find_all("td")[1:]:
        record.append(element.text.strip())
    result.append(record)
    index+=1
        
for record in result:
    print(record)
```

```python
import csv
with open("stats.csv","w") as f:
    writer = csv.writer(f)
    writer.writerow(headers)
    writer.writerows(result)
```
